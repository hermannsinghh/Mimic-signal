```python
# Top signals by strength
if len(df) > 0:
    print("Top 10 signals by strength:")
    top = df.nlargest(10, "strength")[["title", "category", "severity", "confidence", "strength", "source"]]
    display(top)
```

```python
# Signal volume over time
if len(df) > 0:
    df["detected_at"] = pd.to_datetime(df["detected_at"])
    hourly = df.set_index("detected_at").resample("1h").size()
    hourly.plot(figsize=(12, 3), title="Signals per Hour")
    plt.ylabel("Signal count")
    plt.show()
```

```python
# Supply chain vs macro signal cadence
if len(df) > 0:
    pivot = df.groupby([df["detected_at"].dt.date, "category"]).size().unstack(fill_value=0)
    if len(pivot) > 1:
        pivot.plot(kind="bar", stacked=True, figsize=(12, 4), title="Daily Signals by Category")
        plt.show()
    display(pivot)
```
