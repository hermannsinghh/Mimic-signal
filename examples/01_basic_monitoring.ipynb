# Example 01: Basic Signal Monitoring

Start a monitor watching GDELT and SEC 8-K, print every signal above threshold 0.5.

```python
from mimic_signal import SignalMonitor, Signal

monitor = SignalMonitor(threshold=0.5)
monitor.watch(["gdelt", "sec_8k", "fred"])
```

```python
@monitor.on_signal(threshold=0.5)
def handle_signal(signal: Signal, affected_twins: list):
    print(f"\n{'='*60}")
    print(f"SIGNAL DETECTED")
    print(f"  Title:      {signal.title}")
    print(f"  Category:   {signal.category}")
    print(f"  Severity:   {signal.severity:.2f}")
    print(f"  Confidence: {signal.confidence:.2f}")
    print(f"  Strength:   {signal.strength:.2f}")
    print(f"  Source:     {signal.source}")
    print(f"  Sectors:    {signal.affected_sectors}")
    print(f"  Geographies:{signal.affected_geographies}")
    print(f"  URL:        {signal.source_url}")
    print(f"  Detected:   {signal.detected_at.strftime('%Y-%m-%d %H:%M UTC')}")
```

```python
# Non-blocking start — use in Jupyter
import asyncio

task = monitor.start_async()

# Let it run for 60 seconds
await asyncio.sleep(60)

monitor.stop()
```

```python
# Review what came through
df = monitor.signal_log()
print(f"Total signals detected: {len(df)}")
if len(df) > 0:
    display(df[["title", "category", "severity", "confidence", "strength", "source"]]
            .sort_values("strength", ascending=False))
```

```python
# Recent signals from last 24 hours
recent = monitor.recent_signals(hours=24)
print(f"Signals in last 24h: {len(recent)}")
for s in recent:
    print(f"  [{s.severity:.2f}] {s.title[:80]}")
```
